import pandas as pd

# 1. تحديد عنوان URL لصفحة ويكيبيديا الخاصة بـ GameStop
# (هذا هو الرابط الشائع، قد تحتاج للتحقق إذا كانت هناك صفحة أخرى أكثر تفصيلاً للإيرادات)
wikipedia_url_gme = "https://en.wikipedia.org/wiki/GameStop"

# 2. قراءة الجداول من الصفحة
try:
    tables_gme = pd.read_html(wikipedia_url_gme)
except Exception as e:
    print(f"حدث خطأ أثناء قراءة الصفحة: {e}")
    tables_gme = [] # تعيين قائمة فارغة لتجنب الأخطاء إذا فشلت القراءة

# 3. تهيئة متغير الإيرادات لـ GME
gme_revenue = None

# 4. التكرار عبر الجداول المستخرجة والبحث عن الجدول الصحيح لـ GME
for table in tables_gme:
    # معالجة الأعمدة متعددة المستويات لتسهيل البحث
    if isinstance(table.columns, pd.MultiIndex):
        table.columns = [' '.join(col).strip() for col in table.columns]

    # للتحقق من وجود كلمات مفتاحية للتاريخ والإيرادات
    has_date_col_keyword = False
    has_revenue_col_keyword = False

    for col_name in table.columns:
        # البحث عن عمود السنة أو الربع أو الفترة
        if 'Year' in col_name or 'Quarter' in col_name or 'Period' in col_name:
            has_date_col_keyword = True
        # البحث عن عمود الإيرادات (مع كلمات مثل 'Revenue', 'Millions', 'Billions', 'USD')
        if 'Revenue' in col_name and ('Millions' in col_name or 'Billions' in col_name or 'USD' in col_name):
            has_revenue_col_keyword = True
            
    # إذا وجدنا الكلمات المفتاحية في الجدول الحالي، فربما يكون هذا هو جدول الإيرادات
    if has_date_col_keyword and has_revenue_col_keyword:
        gme_revenue = table
        break # بمجرد العثور على الجدول، نخرج من الحلقة

# 5. معالجة وتعديل البيانات إذا تم العثور على جدول الإيرادات
if gme_revenue is not None:
    print("تم العثور على جدول إيرادات GameStop. جاري معالجة البيانات...")

    # تحديد اسم عمود الإيرادات الفعلي بعد المعالجة
    revenue_col_name_gme = None
    for col in gme_revenue.columns:
        if 'Revenue' in col and ('Millions' in col or 'Billions' in col or 'USD' in col):
            revenue_col_name_gme = col
            break
            
    if revenue_col_name_gme:
        # تنظيف عمود الإيرادات: إزالة علامات الدولار والفواصل وتحويلها إلى أرقام
        # تأكد من أن العمود نصي قبل استخدام .str.replace
        gme_revenue[revenue_col_name_gme] = gme_revenue[revenue_col_name_gme].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False)
        
        # إزالة الصفوف التي قد تحتوي على قيم غير رقمية مثل '-' أو '—'
        gme_revenue = gme_revenue[gme_revenue[revenue_col_name_gme] != '-']
        gme_revenue = gme_revenue[gme_revenue[revenue_col_name_gme] != '—'] # لإزالة الشرطات الطويلة أيضاً

        # تحويل عمود الإيرادات إلى نوع بيانات رقمي (عددي)
        # استخدام errors='coerce' لتحويل أي قيم لا يمكن تحويلها إلى NaN، ثم إسقاطها
        gme_revenue[revenue_col_name_gme] = pd.to_numeric(gme_revenue[revenue_col_name_gme], errors='coerce')
        gme_revenue.dropna(subset=[revenue_col_name_gme], inplace=True) # إزالة الصفوف التي بها قيم NaN بعد التحويل

        # 6. عرض الصفوف الخمسة الأخيرة من إطار البيانات (المطلوب في السؤال)
        print("\nالصفوف الخمسة الأخيرة من بيانات إيرادات GameStop:")
        print(gme_revenue.tail())
    else:
        print("خطأ: لم يتم العثور على عمود الإيرادات المناسب في الجدول بعد تحديده.")
else:
    print("لم يتم العثور على جدول إيرادات GameStop في صفحة ويكيبيديا باستخدام الكلمات المفتاحية المحددة.")

# --- الخطوة 2: استيراد المكتبات الضرورية ---
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# --- الخطوة 3: استخراج بيانات أسهم Tesla ---
# رمز المؤشر لشركة Tesla هو "TSLA"
tesla = yf.Ticker("TSLA")
# جلب البيانات التاريخية لأقصى فترة متاحة
tesla_data = tesla.history(period="max")

# إعادة تعيين الفهرس لتحويل التاريخ إلى عمود عادي
tesla_data.reset_index(inplace=True)

# --- الخطوة 4: تعريف دالة make_graph (افتراضية إذا لم تكن متوفرة لديك) ---
# هذه الدالة هي مثال بسيط لكيفية عمل make_graph.
# إذا كانت لديك دالة make_graph معرفة في ملف آخر أو في بيئتك، استخدمها مباشرة.
def make_graph(df, x_col, y_col, title):
    plt.figure(figsize=(10, 6)) # تحديد حجم الرسم البياني
    plt.plot(df[x_col], df[y_col]) # رسم البيانات
    plt.xlabel(x_col) # تسمية المحور السيني
    plt.ylabel(y_col) # تسمية المحور الصادي
    plt.title(title) # عنوان الرسم البياني
    plt.grid(True) # إضافة شبكة للرسم البياني
    plt.show() # عرض الرسم البياني

# --- الخطوة 5: رسم الرسم البياني لأسهم Tesla ---
# استخدام دالة make_graph مع البيانات والعنوان المطلوب
graph_title = "الرسم البياني لأسعار إغلاق أسهم Tesla"
make_graph(tesla_data, 'Date', 'Close', graph_title)

print(f"\nتم رسم الرسم البياني بعنوان: {graph_title}")
print("يرجى التقاط لقطة شاشة للرسم البياني لتقديمها كجزء من الإجابة.")

```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd

# 1. الحصول على بيانات أسهم تسلا (Tesla)
print("جارٍ تنزيل بيانات سهم Tesla (TSLA)...")
tsla_data = yf.download("TSLA", period="5y") # يمكنك تعديل period حسب الحاجة، مثلاً "1y" لسنة واحدة
print("تم تنزيل البيانات بنجاح.")

# 2. تعريف دالة make_graph
def make_graph(stock_data, title):
    """
    تقوم برسم الرسم البياني لسعر إغلاق الأسهم بمرور الوقت.

    المعلمات:
    stock_data (pd.DataFrame): إطار البيانات الذي يحتوي على بيانات الأسهم (يجب أن يحتوي على عمود 'Close').
    title (str): عنوان الرسم البياني.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], color='blue')
    plt.title(title, fontsize=16)
    plt.xlabel("التاريخ", fontsize=12)
    plt.ylabel("سعر الإغلاق (بالدولار الأمريكي)", fontsize=12)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# 3. و 4. استدعاء الدالة make_graph مع البيانات والعنوان المطلوب
graph_title = "الرسم البياني لسعر إغلاق سهم Tesla (TSLA) خلال آخر 5 سنوات"
make_graph(tsla_data, graph_title)
```

```python
# الحصول على بيانات أسهم GameStop (GME)
print("جارٍ تنزيل بيانات سهم GameStop (GME)...")
gme_data = yf.download("GME", period="5y") # يمكنك تعديل period حسب الحاجة
print("تم تنزيل البيانات بنجاح.")

# استدعاء الدالة make_graph مع بيانات GameStop والعنوان المطلوب
graph_title_gme = "الرسم البياني لسعر إغلاق سهم GameStop (GME) خلال آخر 5 سنوات"
make_graph(gme_data, graph_title_gme)

print(f"\nتم رسم الرسم البياني لسهم GameStop بعنوان: {graph_title_gme}")
print("يرجى التقاط لقطة شاشة للرسم البياني لتقديمها.")
```
